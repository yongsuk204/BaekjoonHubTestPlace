***기울림과볼드***

```python
import folium
from IPython.display import display

# 사용자 입력 (쉼표로 구분된 위도,경도 값)
input_data = "37.55540848, 126.9694214, 37.55664444, 126.9736404, 37.55641556, 126.9691925, 37.55677032, 126.9731979, 37.55536652, 126.9686432"

# 입력된 데이터를 파싱하여 리스트로 변환
coords = input_data.split(",")
latitudes = list(map(float, coords[0::2]))  # 짝수 인덱스: 위도
longitudes = list(map(float, coords[1::2]))  # 홀수 인덱스: 경도

# 지도 초기화 (첫 번째 좌표를 중심으로 설정)
map_center = [latitudes[0], longitudes[0]]
m = folium.Map(location=map_center, zoom_start=7)

# 위치 마커 추가
for lat, lon in zip(latitudes, longitudes):
    folium.Marker([lat, lon], popup=f"위도: {lat}, 경도: {lon}").add_to(m)

# Jupyter Notebook에서 지도 출력
display(m)
```
